```python
# Part 1 - Implementation of the Hangman Game

### THIS IS A INST126 HANGMAN PROJECT TRANSLATES ENCRYPTED WORDS 
import random

def decrypt(ciphertext):
    plaintext = ""
    for i in range (0, len(ciphertext)):
        if i % 2 == 0:
            plain_ascii = ciphertext[i]
            cipher_ascii = ord(plain_ascii) - 1
            plaintext = plaintext + str(chr(cipher_ascii))
        else:
            plain_ascii = ciphertext[i]
            cipher_ascii = ord(plain_ascii) + 1
            plaintext = plaintext + str(chr(cipher_ascii))
    return plaintext


print('Welcome to Hangman: INST126 Edition!')

user_input = input("What is your name?: ") 

print('Hello ' + user_input + '!' + ' Try your best and good luck! ')

print('\t1. Play (10 lives)')

game_start = input('Press 1 to start! ')

if (str(game_start) == '1'):
    lives_count = 10  
    print('\nYou have chosen %s and will receive %d lives.' % ('Play', lives_count))
    
words = ['WHDSPQZ', 'XHO', 'TTDBFRT', 'QQJYF', 'ENQD', 'DNPK', 'CNTR', 'EHWHOD',  'TSVMOHOF', 'XNOCFQGTM']

selected_word = random.choice(words)

print('The word contains', len(selected_word), 'letters')

correct = ['_']* len(selected_word)

incorrect = []

def hangman():
    for x in correct:
        print(x, end = ' ')
    print()
hangman()

while True:
    print('\n\n\n') #gives separation between correct and incorrect guesses for less confusion
    guess = input('Please guess a letter: ')
    
    if guess in selected_word:
        index = 0
        for b in selected_word:
            if b == guess:
                correct[index] = guess
                print('Correct')
            index += 1
        hangman() 
    else:
        if guess not in incorrect:
            lives_count -= 1
            incorrect.append(guess)
            print('Incorrect! You have', lives_count,  'lives left')
        else:
            print("You've already guessed that")
            print(incorrect)
    if len(incorrect)>= 10:
        print('Sorry, you lose! Maybe next time!')
        print('I picked', selected_word)
        break   
    if '_' not in correct:
        print('Congratulations, You win! Nice brainpower!')
        print('That word was encrypted.')
        print('The decrypted word is:' ,decrypt(selected_word))
        break
                  
    
    
    

    

    
```
